```python
def getAuth(url, headers={"Authorization":"token {}".format(githubToken)}):
   baseUrl = 'https://api.github.com{}'.format(url)
   print('Asking {}'.format(baseUrl))
   res = requests.get(baseUrl, headers=headers)
   return res.json()

data_lan = getAuth('/repos/ironhack-labs/data-labs/languages')
data1 = getAuth('/repos/ironhack-labs/data-labs/forks?page=1')
data2 = getAuth('/repos/ironhack-labs/data-labs/forks?page=2')
data3 = getAuth('/repos/ironhack-labs/data-labs/forks?page=3')
data4 = getAuth('/repos/ironhack-labs/data-labs/forks?page=4')
print(data_lan.keys())
```
